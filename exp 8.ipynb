```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
train=pd.read_csv('/content/drive/MyDrive/data science/youtube.csv')
test=pd.read_csv('/content/drive/MyDrive/data science/test.csv')
```

```python
train=train.dropna()
print(train.shape)
print(list(train.columns))
```

```python
test=test.dropna()
print(test.shape)
print(list(test.columns))
```

```python
x=np.array(train['likes']).reshape((-1,1))
print(x)
```

```python
y=np.array(test['likes'])
print(y)
```

```python
model=LinearRegression()
model=LinearRegression().fit(x, y)
```

```python
r_sq=model.score(x,y)
print('coefficient of determination:', r_sq)
print('intercept :', model.intercept_)
print('slope:', model.coef_)
```

```python
y_pred=model.predict(x)
print('predicted response:' , y_pred, sep='\n')
```

```python
df=pd.DataFrame({'Actual': y_pred.flatten(), 'Predicted': y.flatten()})
df
```

```python
result=df.tail(30)
print(result)
```

```python
result.plot(kind='bar',figsize=(16,10))
plt.grid(which='major',linestyle='-',linewidth='0.5', color='green')
plt.grid(which='minor',linestyle=':',linewidth='0.5', color='black')
plt.show()
```
